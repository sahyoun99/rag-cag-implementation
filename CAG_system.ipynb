# Load pdf from drive

```python
from langchain_community.document_loaders import PyPDFLoader

# Replace '/content/drive/MyDrive/UAE Constitution /The Constitution of the United Arab Emirates.pdf' with the actual path to your PDF file in Google Drive
pdf_path = "/content/drive/MyDrive/UAE Constitution /The Constitution of the United Arab Emirates.pdf"

loader = PyPDFLoader(pdf_path)
documents = loader.load()

# Combine all pages into ONE context (CAG style)
document_context = "\n\n".join([doc.page_content for doc in documents])
```

# Create LLM

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)
```

# Ask Questions using the CAG Chain

```python
SYSTEM_PROMPT = f"""
You are an assistant that answers questions strictly using the following document.

DOCUMENT:
{document_context}

If the answer is not in the document, say "The document does not contain this information."
"""
```

```python
print("ðŸ“˜ Document Q&A Chat (CAG Mode)")
print("Type your question or 'exit' to quit.\n")

while True:
    user_input = input("You: ").strip()

    if user_input.lower() in {"exit", "quit", "q"}:
        print("ðŸ‘‹ Goodbye!")
        break

    if not user_input:
        continue

    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": user_input}
    ]

    response = llm.invoke(messages)

    print("\nðŸ¤– Answer:")
    print(response.content)
    print("-" * 60)
```
